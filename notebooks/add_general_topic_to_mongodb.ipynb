```python
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
with open('my_export_en.json') as f:
    data = json.load(f)
```

```python
topic = 'general'
language = 'english'
def add_posts(data, language, topic, Id=1, old_Id=1):
    posts = []
    for i in range(len(data['conversations'])):
        for j in range(2):
            if j==0:
                PostTypeId = 1  
                post = {
                    'Id' : Id,
                    'PostTypeId' : str(PostTypeId), 
                    'Score' : '1',
                    "ViewCount" : "1",
                    'Title' : data['conversations'][i][j],
                    "AnswerCount" : "1",
                    'Language' : language,
                    'Topic' : topic,
                }
            else:
                PostTypeId = 2
                post = {
                    'Id' : Id,
                    'PostTypeId' : str(PostTypeId),
                    'ParentId' : str(old_Id),
                    'Score' : '1',
                    'Body' : data['conversations'][i][j],
                    'Language' : language,
                    'Topic' : topic,
                }
            posts.append(post)
            old_Id = Id
            Id +=1
    return posts, Id, old_Id
```

```python
posts, Id, old_Id = add_posts(data, language, topic)
posts
```

```python
import pymongo
import json
from pymongo import MongoClient
```

```python
connection = MongoClient()
db = connection["StarkBotBD"]
Quest_Rep = db["Quest_Rep"]
db.Quest_Rep.insert(posts)
```

```python
with open('my_export_fr.json') as f:
    data = json.load(f)

topic = 'general'
language = 'french'
posts, Id, old_Id = add_posts(data, language, topic, Id+1, old_Id+1)
```
